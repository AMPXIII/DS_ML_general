```python
df_m = df[['age','checking_status','duration','job','purpose','accepted']]
df_m.head()
```

```python
df_m.info()
```

```python
df_m.purpose = df_m.purpose.astype('category')
df_m.checking_status = df_m.checking_status.astype('category')
df_m.job = df_m.job.astype('category')
```

```python
df_m.info()
```

```python
df_m_dummified = pd.get_dummies(df_m, columns = ['checking_status','job','purpose'], drop_first=True)
```

```python
df_m_dummified.head()
```

```python
X = df_m_dummified.drop('accepted', axis=1)
y = df_m_dummified['accepted']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_clf = LogisticRegression()
```

```python
lr_clf.fit(X,y)
```

```python
lr_clf.score(X,y)
```

```python
from sklearn.metrics import recall_score, precision_score
```

```python
print('Recall: ',recall_score(y, lr_clf.predict(X)))
print('Precision: ',precision_score(y, lr_clf.predict(X)))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(X)
```

```python
X_std = sc.transform(X)
```

```python
lr_clf_std = LogisticRegression()
lr_clf_std.fit(X_std,y)
print('Accuracy: ', lr_clf_std.score(X_std,y))
print('Recall: ',recall_score(y, lr_clf_std.predict(X_std)))
print('Precision: ',precision_score(y, lr_clf_std.predict(X_std)))
```

```python
import pandas as pd
import random
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from statistics import mean, stdev
from math import exp
```

```python
%matplotlib inline
```

```python
# Checking the importance of a coefficient by computing its "Beta"
data = pd.read_csv("loan_data_track1.csv")
numeric_model_fields = ["age", "duration", "loan_amount", "installment_commitment"]
X = data[numeric_model_fields]
X.insert(len(X.columns), 'own_telephone_Yes', pd.get_dummies(data['own_telephone'])['Yes'])
X_col_list = list(X.columns)
y = data["accepted"]
logistic_model = LogisticRegression()
sample_size = 0.6
result_list = []
for i in range(100):
    sample_index = random.sample(range(len(data)),int(len(data) * sample_size))
    model = logistic_model.fit(X.ix[sample_index], y.ix[sample_index])
    result_list.append(exp(model.coef_[0][X_col_list.index('own_telephone_Yes')]))
    
print(mean(result_list))
print(stdev(result_list))
plt.hist(result_list)
```
